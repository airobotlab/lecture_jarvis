```python
import pyttsx3
def 스피커(text='자비스 안녕? 준비됐어?', 목소리선택=0):

    engine = pyttsx3.init()
    voices = engine.getProperty('voices')
    engine.setProperty('voice', voices[목소리선택].id)

    engine.say(text)
    engine.runAndWait()
    return

스피커(text='자비스 안녕? 준비됐어?', 목소리선택=0)
스피커(text='Hello Jarvis?? Are you ready??', 목소리선택=1)
스피커(text='Hello Jarvis?? Are you ready??', 목소리선택=2)
```

```python
## 시간 함수
import datetime

def 시간():
    현재시간 = datetime.datetime.now().strftime('%I:%M')
    text = f'현재 시간은 {현재시간} 입니다.'
    print(text)
    스피커(text=text)
    return

시간()
```

```python
## 날짜 함수
def 날짜():
    
    오늘_몇년 = str(datetime.datetime.now().year)
    오늘_몇월 = str(datetime.datetime.now().month)
    오늘_며칠 = str(datetime.datetime.now().day)
    
    text = f'오늘은 {오늘_몇년}년 {오늘_몇월}월 {오늘_며칠}일 입니다.'
    print(text)
    스피커(text=text)
    return

날짜()
```

```python
## 인사 함수
def 인사(이름='우영'):
    현재_시 = int(datetime.datetime.now().hour)
    if 현재_시 >=6 and 현재_시 < 12:
        print(f'아침 {현재_시} 시')
        스피커(text=f'Good morning {이름}')
    elif 현재_시 >= 12 and 현재_시 < 18:
        print(f'오후 {현재_시} 시')
        스피커(text=f'Good afternoon {이름}')
    elif 현재_시 >= 18 and 현재_시 <= 24:
        print(f'저녁 {현재_시} 시')
        스피커(text=f'Good evening {이름}')
    else:
        print(f'새벽 {현재_시} 시')
        스피커(text=f'Good night {이름}')        
    return

인사(이름=이름)
```

```python
## 자비스 부르기 함수
def 첫인사(이름='우영'):
    스피커(text='환영합니다')
    인사(이름=이름)
    날짜()
    시간()
    스피커(text='자비스 준비됐습니다, 도와드릴일 있으신가요?')
    return

첫인사(이름=이름)
```

```python
## 뉴스 크롤링 함수
from bs4 import BeautifulSoup
import requests

def 네이버뉴스검색함수(query='블랙핑크', 개수=3):

#     query = '블랙핑크'  # 검색어
    page = 0  # 크롤링할 페이지
    page_num = 0  #start수를 1, 11, 21, 31 ...만들어 주는 함수
    if page == 1:
        page_num =1
    elif page == 0:
        page_num =1
    else:
        page_num = page+9*(page-1)

    #url 생성
    url = f'https://search.naver.com/search.naver?where=news&sm=tab_pge&query={query}&start={str(page_num)}'
    print("네이버 검색 URL: %s"%url)

    # ConnectionError방지
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) Chrome/100.0.48496.75"}

    #html불러오기
    original_html = requests.get(url, headers=headers)
    html = BeautifulSoup(original_html.text, "html.parser")

    # 검색결과
    articles = html.select("div.group_news > ul.list_news > li div.news_area > a")
    print(len(articles),"개의 기사가 검색됌.")  # 검색된 기사의 갯수

    #뉴스기사 제목 가져오기
    news_title = []
    for i in articles:
        news_title.append(i.attrs['title'])
    
    뉴스제목들 = news_title[:개수]
    
    text = f'{query} 검색결과는\n  {뉴스제목들[0]}\n  {뉴스제목들[1]}\n  {뉴스제목들[2]}'
    print('## '+ text)
    
    return text

text = 네이버뉴스검색함수(query='블랙핑크')
```
